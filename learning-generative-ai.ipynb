## Generate text from text inputs

For text-only prompts, use the gemini-pro model:

```python
model = genai.GenerativeModel('gemini-pro')
```

The **generate_content** method can handle a wide variety of use cases, including multi-turn chat and multimodal input, depending on what the underlying model supports. The available models only support text and images as input, and text as output.

```python
%%time
response = model.generate_content("What is the meaning of life?")
```

```python
response.text
```

```python
for chunk in response:
  print(chunk.text)
  print("_"*80)
```

If the API failed to return a result, use **prompt_feedback** to see if it was blocked due to safety concerns regarding the prompt.

```python
response.prompt_feedback
```

Gemini can generate multiple possible responses for a single prompt. These possible responses are called **candidates**, and you can review them to select the most suitable one as the response.

```python
response.candidates
```

# Langchain

LangChain is an open-source framework designed to simplify the creation of applications using large language models (LLMs). It provides a standard interface for chains, lots of integrations with other tools, and end-to-end chains for common applications.

```python
pip install langchain
```

## Prompt Templates

```python
from langchain.prompts import PromptTemplate
```

```python
prompt_template_name=PromptTemplate(
    input_variables=['country'],
    template='Can you tell me the capital of {country}'
)
```

```python
prompt_template_name.format(country='Bangladesh')
```

```python
pmt1 = prompt_template_name.format(country='Bangladesh')
```

```python
pmt2 = prompt_template_name.format(country='China')
```

```python
response = model.generate_content(pmt1)
```

```python
response.text
```

```python
response = model.generate_content(pmt2)
```

```python
response.text
```

```python
prompt_template_name = PromptTemplate.from_template('What is a good name for a company that makes a {product}')
```

```python
prompt_template_name.format(product='toy')
```
